## 复制原有data到clean data，之后保存clean data，不动原始的dataframe

```python
clean_data = original_data.copy()
```

## 把null的行删除

```python
clean_data[clean_data["Aging"].isnull()]
```

```python
clean_data.dropna(subset=["Aging"],inplace=True)
clean_data[clean_data["Aging"].isnull()]
```

```python
clean_data[clean_data["Sales"].isnull()]
```

```python
clean_data.dropna(subset=["Sales"],inplace=True)
clean_data["Sales"].isnull().sum()
```

```python
clean_data.dropna(subset=["Quantity","Discount","Shipping_Cost","Order_Priority"],inplace=True)
```

```python
clean_data.info()
```

```python
clean_data[clean_data["Profit"].isnull()]
```

```python
clean_data.set_index("Customer_Id",inplace=True)
clean_data.head()
```

```python
clean_data.reset_index(inplace=True)
clean_data.head()
```

## 检查是否有重复数据。这里customer id可以被重复，一个人可能重复购买。我们这里只是为了演示删除重复数据的操作

```python
clean_data[clean_data["Customer_Id"].duplicated()]
clean_data.drop_duplicates(subset=["Customer_Id"],inplace=True)
clean_data[clean_data["Customer_Id"].duplicated()]
clean_data.info()
```

## 更改数据结构

```python
clean_data["Order_Date"] = pd.to_datetime(clean_data["Order_Date"])
clean_data["Order_Date"]
```

### 这里为了验证slice用法，可以切除字符串的一部分

```python
clean_data["Customer_Id"] = clean_data["Customer_Id"].astype(str) 
clean_data["Customer_Id"] = clean_data["Customer_Id"].str.slice(0,4)
clean_data["Customer_Id"]
```

```python
clean_data["Gender"] = clean_data["Gender"].astype("category")
clean_data["Gender"]
```

```python
clean_data["Device_Type"] = clean_data["Device_Type"].astype("category")
clean_data["Device_Type"]
```

```python
clean_data["Customer_Login_type"] = clean_data["Customer_Login_type"].astype("category")
clean_data["Customer_Login_type"] 
```

```python
clean_data["Order_Priority"] = clean_data["Order_Priority"].astype("category")
clean_data["Order_Priority"]
```

```python
clean_data["Payment_method"] = clean_data["Payment_method"].astype("category")
clean_data["Payment_method"] 
```

```python
clean_data["Product"] = clean_data["Product"].replace("Suits","Jeans") #replace(["Suits","Shirts"],"Jeans") , replace({"Suits":"Jeans","LCD":"Watch"})
clean_data["Product"].value_counts() #通过两种方法check是否Suits已经被替换了
len(clean_data[clean_data["Product"] == "Suits"]) #通过两种方法check是否Suits已经被替换了
```
